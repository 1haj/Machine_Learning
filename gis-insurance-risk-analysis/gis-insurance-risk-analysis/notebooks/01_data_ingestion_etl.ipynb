```python
import pandas as pd

claims = pd.read_csv("../data/raw/insurance_claims_1000.csv")
regions = pd.read_csv("../data/raw/regions.csv")

# Basic cleaning
claims.dropna(inplace=True)

# Feature engineering
claims["incident_date"] = pd.to_datetime(claims["incident_date"])
claims["high_claim"] = (claims["claim_amount"] > 1500).astype(int)

# Save processed data
claims.to_csv("../data/processed/claims_processed.csv", index=False)

claims.head()
```
